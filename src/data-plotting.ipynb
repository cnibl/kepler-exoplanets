The plot below shows the position on the night sky where the objects are, in terms of right ascension (`ra`) and declination (`dec`). The observations come from a fairly small portion compared to the whole sky (about 15 times 20 degrees).

```python
jp = sns.jointplot(x="ra", y="dec", data=kepler_data, kind="hex", space=0)
```

# Classification
The two plots below show the classification of the observed objects, `koi_pdisposition` is the classification from the Kepler data alone whereas `koi_disposition` also includes information from the Exoplanet Archive (an exoplanet database), with the Kepler classification used for objects which are not deemed `CONFIRMED`. 

`CONFIRMED` objects are in the Exoplanet Archive designated as confirmed exoplanetary objects. `CANDIDATE` objects are those observed by Kepler that have passed a number of tests to determine whether it is a suitable candidate, `FALSE POSITIVE` objects have failed at least one of these tests.

```python
sns.countplot(x="koi_disposition", data=kepler_data)
```

```python
sns.countplot(x="koi_pdisposition", data=kepler_data)
```

# Identifying correlating variables
To find which of the many variables that are mostly correlated with the classification of an object (its `koi_pdisposition` and `koi_disposition` values), we look at the correlation matrix between the different variables. We drop unnecessary columns with names and measurement errors.

```python
kepler_reduced = kepler_data.drop(['rowid', 'kepid', 'kepoi_name', 'kepler_name', 'koi_tce_delivname', 
                     'koi_period_err1', 'koi_period_err2', 
                     'koi_time0bk_err1', 'koi_time0bk_err2', 
                     'koi_impact_err1', 'koi_impact_err2',
                     'koi_depth_err1', 'koi_depth_err2', 
                     'koi_prad_err1', 'koi_prad_err2', 
                     'koi_insol_err1', 'koi_insol_err2',
                     'koi_steff_err1', 'koi_steff_err2', 
                     'koi_slogg_err1', 'koi_slogg_err2', 
                     'koi_srad_err1', 'koi_srad_err2', 
                     'koi_duration_err1', 'koi_duration_err2', 
                     'koi_teq_err1', 'koi_teq_err2'], axis=1)
kepler_reduced = kepler_reduced.replace({"CONFIRMED": 0, "CANDIDATE": 1, "FALSE POSITIVE": 2})
corr_matrix = kepler_reduced.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(abs(corr_matrix), cmap="binary", square=True)
abs(corr_matrix[["koi_disposition", "koi_pdisposition"]])
```

## Most correlating variables
The variables that are most correlated are the ones in the below heatmap. They should be most relevant for determining how an object should be classified.

```python
kepler_mostcorr = kepler_reduced.drop(["koi_period", "koi_time0bk", "koi_impact", "koi_duration",
                                       "koi_prad", "koi_insol", "koi_steff", "koi_srad",
                                       "ra", "dec", "koi_kepmag", "koi_slogg"], axis=1)
print(kepler_mostcorr.columns)
sns.heatmap(abs(kepler_mostcorr.corr()), cmap="binary")
```

# Pairwise scatter plots
Below are scatter plots showing the separations between the classes for the most correlating variables of `float` type (the `fpflag` variables are discrete with value of 0 or 1). We see that the separation between the classes is not too great, which will mean that even if these are variables that correlate strongly with the classification, we run a risk of not getting a very good classification accuracy in the end.

```python
plot_cols = ["koi_depth","koi_teq","koi_model_snr","koi_tce_plnt_num"]
kepler_mostcorr[plot_cols].dtypes
no_nan = kepler_mostcorr.copy()
no_nan.fillna(kepler_mostcorr.mean(), inplace=True)
sns.pairplot(x_vars=plot_cols, y_vars=plot_cols, data=no_nan, hue="koi_disposition", hue_order=[2,1,0], diag_kind="kde")
```

# Conclusion of analysis of data variables
With the above in mind, it is clear that a classification analysis aiming at predicting whether an object is a `CANDIDATE` or a `FALSE POSITIVE` should primarily use the columns with the most correlating variables. Not surprisingly, the `fp_flag` variables, that designate whether an object has passed the false positive tests are important.
